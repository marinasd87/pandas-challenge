```python
import pandas as pd
```

```python
#File to load
file_to_load = "/Users/marinaduarte/Documents/nu-chi-data-pt-09-2020-u-c/04-Pandas/HW/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"
#Read purchasing file and store into pandas datafram
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

```python
print("Number of players: ", purchase_data['SN'].nunique(dropna=True))
```

```python
number_of_players = purchase_data['SN'].nunique(dropna=True)
number_of_playersDF = pd.DataFrame({"Number Of Players": [number_of_players]})
number_of_playersDF
```

```python
number_of_unique_items = purchase_data['Item Name'].nunique(dropna=True)
average_price = purchase_data['Price'].mean()
number_of_purchases = purchase_data['Purchase ID'].count()
total_revenue = purchase_data['Price'].sum()
purchasing_analysis = pd.DataFrame({"Number of Unique Items": [number_of_unique_items],
                                   "Average Price": [average_price],
                                   "Number of purchases": [number_of_purchases],
                                   "Total Revenue": [total_revenue]})
purchasing_analysis
```

```python
df2 = purchase_data.loc[:,['SN','Gender']]
df3 = pd.DataFrame.drop_duplicates(df2)
df3
```

```python
total_count = df3.groupby(['Gender'])[['SN']].count()
total_count
```

```python
total_count['percentage'] = (total_count['SN']/total_count['SN'].sum())*100
print(total_count)
```

```python
total_countDF = pd.DataFrame(total_count)
total_countDF = total_countDF.rename(columns={"SN": "Total Count", "percentage": "Percentage Of Players"})
total_countDF['Percentage Of Players'] = pd.Series(["{0:.2f}%".format(val) for val in total_countDF['Percentage Of Players']], index = total_countDF.index)
total_countDF
```

```python
pa_gender4 = purchase_data.groupby(['Gender'])
pa_gender_PC = pa_gender4['Price'].count()
pa_gender_AP = pa_gender4['Price'].mean()
pa_gender_TPV = pa_gender4['Price'].sum()
pa_gender_person = total_countDF['Total Count']
pa_gender_ATPP = pa_gender_TPV / pa_gender_person
pa_gender_final = pd.DataFrame({"Purchase Count": pa_gender_PC,
                               "Average Purchase Price": pa_gender_AP,
                               "Total Purchase Value": pa_gender_TPV,
                               "Avg Total Purchase per Person": pa_gender_ATPP})
pa_gender_final
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data.head()
```

```python
age_group = purchase_data.loc[:,['SN','Age Group']]
age_group1 = pd.DataFrame.drop_duplicates(age_group)
ageDF = age_group1.groupby(['Age Group'])[['SN']].count()
ageDF['Percentage of Players'] = (ageDF['SN']/ageDF['SN'].sum())*100
ageDF = ageDF.rename(columns={"SN": "Total Count"})
ageDF
```

```python
pa_age = purchase_data.groupby(['Age Group'])
pa_age_PC = pa_age['Price'].count()
pa_age_AP = pa_age['Price'].mean()
pa_age_TPV = pa_age['Price'].sum()
pa_age_person = ageDF['Total Count']
pa_age_ATPP = pa_age_TPV / pa_age_person
pa_age_final = pd.DataFrame({"Purchase Count": pa_age_PC,
                               "Average Purchase Price": pa_age_AP,
                               "Total Purchase Value": pa_age_TPV,
                               "Avg Total Purchase per Person": pa_age_ATPP})
pa_age_final
```

```python
top_spenders = purchase_data.groupby(['SN'])
top_spenders_PC = top_spenders['Price'].count()
top_spenders_AP = top_spenders['Price'].mean()
top_spenders_TPV = top_spenders['Price'].sum()
top_spenders_final = pd.DataFrame({"Purchase Count": top_spenders_PC,
                               "Average Purchase Price": top_spenders_AP,
                               "Total Purchase Value": top_spenders_TPV})
top_spenders_final = top_spenders_final.sort_values(by=['Total Purchase Value'], ascending=False)
top_spenders_final.head()
#top_spenders_final.nlargest(5,'Total Purchase Value')
```

```python
item_group = purchase_data.loc[:,['Item ID','Item Name','Price']]
top_items = item_group.groupby(['Item ID','Item Name'])
top_items_PC = top_items['Price'].count()
top_items_AP = top_items['Price'].mean()
top_items_TPV = top_items['Price'].sum()
top_items_final = pd.DataFrame({"Purchase Count": top_items_PC,
                               "Item Price": top_items_AP,
                               "Total Purchase Value": top_items_TPV})
top_items_final = top_items_final.sort_values(by=['Purchase Count'], ascending=False)
top_items_final.head()
```

```python
most_profitable_items = top_items_final.sort_values(by=['Total Purchase Value'], ascending=False)
top_items_final.head()
```
